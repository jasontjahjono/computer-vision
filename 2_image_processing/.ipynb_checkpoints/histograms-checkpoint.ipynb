```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('../DATA/horse.jpg') # ORIGINAL BGR FOR OPENCV
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('../DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('../DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
# OPENCV BGR
hist_values = cv2.calcHist([blue_bricks], channels=[0], mask=None,
                           histSize=[256], ranges=[0, 256])
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
# OPENCV BGR
hist_values = cv2.calcHist([dark_horse], channels=[0], mask=None,
                           histSize=[256], ranges=[0, 256])
plt.plot(hist_values)
```

```python
img = blue_bricks
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
plt.title("HISTOGRAM FOR BLUE BRICKS")
```

```python
img = dark_horse
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])
plt.title("HISTOGRAM FOR HORSE")
```
