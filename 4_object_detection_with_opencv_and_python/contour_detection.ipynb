```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/internal_external.png', 0)
img.shape
```

```python
plt.imshow(img, cmap="gray")
```

```python
# openCV allows us to find internal and external contours

image, contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
# contours = number of internal and external contours
# hierarchy = np.ndarray of lines where contours are found
```

```python
external_contours = np.zeros(image.shape)
```

```python
for i in range(len(contours)):
    
    # External Contour
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours, contours, i, 255, -1)
```

```python
plt.imshow(external_contours, cmap="gray")
```

```python
internal_contours = np.ones(image.shape) * 255
for i in range(len(contours)):
    # Internal Contour
    if not hierarchy[0][i][3] == -1:
        cv2.drawContours(internal_contours, contours, i, 0, -1)
plt.imshow(internal_contours, cmap='gray')
```

```python
pepperoni_contours = np.ones(image.shape) * 255
for i in range(len(contours)):
    # Internal Contour
    if hierarchy[0][i][3] == 4:
        cv2.drawContours(pepperoni_contours, contours, i, 0, -1)
plt.imshow(pepperoni_contours, cmap='gray')
```

```python
smiley_contours = np.ones(image.shape) * 255
for i in range(len(contours)):
    # Internal Contour
    if hierarchy[0][i][3] == 0:
        cv2.drawContours(smiley_contours, contours, i, 0, -1)
plt.imshow(smiley_contours, cmap='gray')
```
